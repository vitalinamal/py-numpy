### Task 7.2

````python
# Python code
def unknown_signature(lst1, lst2):
    bitwise_complement = [~x for x in lst1]
    bitwise_and = [x & y for x, y in zip(lst1, lst2)]
    bitwise_or = [x | y for x, y in zip(lst1, lst2)]
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
print("Bitwise complement:", result_bitwise_complement)
print("Bitwise AND:", result_bitwise_and)
print("Bitwise OR:", result_bitwise_or)
````

````python
def unknown_signature(lst1, lst2):
    lst1 = np.array(lst1)
    lst2 = np.array(lst2)
    bitwise_complement = ~lst1
    bitwise_and = lst1 & lst2
    bitwise_or = lst1 | lst2
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
print("Bitwise complement:", result_bitwise_complement)
print("Bitwise AND:", result_bitwise_and)
print("Bitwise OR:", result_bitwise_or)
````
